***Numpy Fundamentals***

>1. Creating Numpy Array

```python
import numpy as np

lst = [10,20,30,40,50]
arr = np.array([1,2,3,4,5])
print(lst, type(lst))
print(arr, type (arr))
```

>2. Creating Basic Arrays

```python
# 1D array -> Vectors
a = np.array([1,2,3,4])
print(a)

# 2D array -> Metrix
b = np.array([ [1,2,3,4] ])
print(b)

# 3D array -> Tensors
c = np.array([ [[10,20, 30, 40], [30,40,50,60]] ])
print(c)
```

>3. Specifying data types of numpy array

```python
# Creating floating arrays
a = np.array([10,20,30,40], dtype=float)
print(a, a.dtype)

# Creating Boolean arrays
b = np.array([0, 10, -10, 90, 66, 10.6], dtype=bool)
print(b, b.dtype)
```

>4. Array Initialization Functions

```python
# 1) np.arange()
a  = np.arange(1,10)   # last value is exclusive
print(a)

b  = np.arange(2,21, 2)   # -> even  number of np array
print(b)

# 2) np.ones
ones_array = np.ones( (3,2) )
print(ones_array)
print()

# 3) np.zeros
zeros_array = np.zeros( (3,2) )
print(zeros_array)

# 4) np.random
random_array = np.random.random( (3,2) )  # generates num b/w 0 and 1
print(random_array, 1)

# 5) linspace
l = np.linspace(1,5, 10)
print(l)

# 6) Identity metrix
i = np.identity(3)
print(i)

# 💡 Note -> In case of ones, zeros & random we pass tuple
```

>7. Numpy Array Attributes

```python
# 1) ndim -> It runs the dimentions of an array.

# Vector
A1 = np.arange(1,11)
print(A1, '->', A1.ndim)

# Metrix
A2 = np.arange(1,13).reshape(3,4)
print(A2, '->', A2.ndim)

# Tensor
A3 = np.arange(1,19).reshape(3,2,3)
print(A3, '->', A3.ndim)
```

```python
# 3) shape --> It simply returns a tuple which specifies the total number of elements

# Vector
A1 = np.arange(1,11)
print(A1, '->', A1.shape)

# Metrix
A2 = np.arange(1,13).reshape(3,4)
print(A2, '->', A2.shape)

# Tensor
A3 = np.arange(1,19).reshape(3,2,3)
print(A3, '->', A3.shape)
```

```python
# 4) size  -> It returns the total number of element in array.

# Vector
A1 = np.arange(1,11)
print(A1, '->', A1.size)

# Metrix
A2 = np.arange(1,13).reshape(3,4)
print(A2, '->', A2.size)

# Tensor
A3 = np.arange(1,19).reshape(3,2,3)
print(A3, '->', A3.size)
```

```python
# 5) Item size  --> it returns the size of  individual elements of an array in bytes 

# Vector
A1 = np.arange(1,11)  # dtype -> int 64 by default
print(A1, '->', A1.itemsize)

# Metrix
A2 = np.arange(1,13, dtype='int32').reshape(3,4)
print(A2, '->', A2.itemsize)

# Tensor
A3 = np.arange(1,19, dtype='int16').reshape(3,2,3)
print(A3, '->', A3.itemsize)
```

```python
# 6) dtype  --> it tells the data type of an array

# Vector
A1 = np.arange(1,11) 
print(A1, '->', A1.dtype)

# Metrix
A2 = np.arange(1,13, dtype='int32').reshape(3,4)
print(A2, '->', A2.dtype)

# Tensor
A3 = np.arange(1,19, dtype='int16').reshape(3,2,3)
print(A3, '->', A3.dtype)
```

>8. Array Operation - ( Scalar & Vector )

```python
# 1. Scalar Operations:

# Arithmetic --> + , - , *, / , //, %, ** 
m = np.arange(1,13).reshape(3,4)
print(m,'\n')
result = m + 2
print(result) 

# Relational --> > , < , >=, <= , ==, != 
r = np.arange(1,13).reshape(3,4)
print(r,'\n')
result = r > 9
print(result) 
```

```python
# 2. Vector Operations.
a1 = np.arange(1,16).reshape(3,5)
print(a1)
print('\n')

a2 = np.arange(10,25).reshape(3,5)
print(a2)
print('\n')

res = a1 + a2
print(res)
```

>9. Array Functions:


```python
# max, min, sum, prod
a = np.arange(1,21).reshape(5,4)
print(a)
print(f'Max = {np.max(a)}') 
print(f'Min = {np.min(a)}') 
print(f'Sum = {np.sum(a)}') 
print(f'Prod = {np.prod(a)}') 
```

```python
# Axis Parameter
a = np.arange(1,21).reshape(5,4)
print(a)
print(f'Max of each rows = {np.max(a, axis=1)}')
print(f'Min of each column = {np.min(a, axis=0)}')
print(f'Sum of each row = {np.sum(a, axis=1)}')
```

```python
# Statistical Functions
b = np.arange(10,30).reshape(5,4)
print(b)
print(f'Mean = {np.mean(b)}')
print(f'Median = {np.median(b)}')
print(f'Standard Deviation = {np.std(b)}')
print(f'Variance = {np.var(b)}')
```

```python
# Dot product
a1 = np.arange(1,13).reshape(3,4)
print(a1)
print()

a2 = np.arange(0,12).reshape(4,3)
print(a2)
print()

result = np.dot(a1,a2)
print(result)
```

```python
# log and exponent 

a = np.arange(10,30).reshape(5,4)
print(a) 
print() 

log_metrix = np.log(a)
print(log_metrix)
print()

exp_metrix = np.exp(a)
print(exp_metrix)
print()
```

```python
# round, floor, ceil
b = np.random.random((5,5))*100
print(b) 
print() 
print( np.round(b)) 
```

>10. Indexing
- Positive indexing
- Negative indexing

```python
# 1) Vector Indexing
a = np.array([10,20, 30, 40, 50, 60])
print(a)

print(f"First element: {a[0]}")    # -> positive indexing   
print(f"Last element: {a[-1]}")    # -> negative indexing
```

>11. Slicing

```python
# 1) Slicing vectors
a = np.arange(10)
print(a)

#  Slice -> 0, [1,2,3], [7,8,9], [1,3,5,7], [2,4,6,8], [3,6] 
print( a[0:1] )
print( a[1:4] )
print( a[-3:] )
print( a[1:8:2] )
print( a[2:9:2] )
print( a[3:7:3] )
```

```python
# 2) Slicing Metrix
b = np.arange(12).reshape(3,4)
print(b)
print()

# slice -> 1st row
print (b[0, ::] )

# slice -> 3rd row
print (b[2, ::] )

# slice -> 2nd column
print (b[::, 2] )
```

```python
# Slicing sub metrix:
''' Questions:
  i) [ 5 6 ]              
     [ 9 10 ]

  ii) [ 6 7 ]
      [ 10 11] 

  iii) [ 2 3 ]
       [ 6 7] 

   iv) [ 4 5 ]
       [ 8 9 ] 

   v) [ 2 3 ]
      [ 10 11 ] 

   vi)  [ 0 3 ]
        [ 8 11 ] 
'''  
m = np.arange(12).reshape(3,4)
print(m)
print()

# i)
print( m[1: ,1:3 ] )
print()

# ii)
print( m[1: ,2:4 ] )
print()

# iii)
print( m[:2 ,2: ] )
print()

# iv)
print( m[1: , :2 ] )
print()

# v)
print( m[::2 , 2:] )
print()

# vi)
print( m[::2 , ::3] )
print()

# vii) [1,3,9,11]
print( m[::2 , 1::2] )
print()
```

>12. Iteration on numpy array

```python
# 1) Iteration on vectors
a = np.arange(10)

for e in a:
  print(e, end=' ')
```

```python
# 3) Iteration on tensors
a = np.arange(18).reshape(3,3,2)

for e in a:
  print(e)
```

>13. Advance iteration on numpy array -> nditer()

```python
# Iteration on tensors
a = np.arange(18).reshape(3,3,2)

for e in np.nditer(a):
  print(e, end=' ')
```

>16. Numpy Stacking

```python
# Horizontal Stacking
a1 = np.arange(12).reshape(3,4)
print(a1)
print()

a2 = np.arange(15,27).reshape(3,4)
print(a2)
print()

h_stack = np.hstack((a1, a2, a1))
print(h_stack)
```

```python
# Vertical Stacking
a1 = np.arange(12).reshape(3,4)
print(a1)
print()

a2 = np.arange(15,27).reshape(3,4)
print(a2)
print()

h_stack = np.vstack((a1, a2))
print(h_stack)
```
