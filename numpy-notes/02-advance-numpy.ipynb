>1. Numpy array vs Python list

```python
# Comparision based on time -> Python list
import time

start = time.time()
l1 = [e for e in range(10000000)]
l2 = [e for e in range(10000000)]

result = []

for e in range(len(l1)):
  result.append( l1[e] + l2[e] )

print(f'Time taken by list = {time.time() - start} sec')
```

```python
# Comparision based on time -> Numpy array
import numpy as np
import time

start  = time.time()
a1 = np.arange(10000000)
a2 = np.arange(10000000)

res = a1 + a2
print(f'Time taken numpy array = {time.time() - start} sec')
```

```python
# Comparision based on memory -> Python list
import sys

l1 = [e for e in range(10000000)]
print(f'List size in bytes = {sys.getsizeof(l1)} bytes')
```

```python
# Comparision based on memory -> Numpu array
import sys

a = np.arange(10000000, dtype='int16')
print(f'Numpy array size in bytes = {a.nbytes} bytes')
```

>4. Broadcasting

```python
# Same dimension but different shape => Broadcasting ✅
a = np.arange(6).reshape(2,3)
b = np.arange(3).reshape(1,3)

print(a + b)
```

```python
# (4x3) Metrix and (3,) vector => Broadcasting ✅
a = np.arange(12).reshape(4,3)
b = np.arange(3)

print(a + b)
```

```python
# Different dimension and there is no 1 => Broadcasting ❌
a = np.arange(12).reshape(3,4)
b = np.arange(3)

print(a + b)
```

```python
# (1x3) and (3x1) => Broadcasting ✅
a = np.arange(3).reshape(1,3)
b = np.arange(3).reshape(3,1)

print(a + b)
```

```python
# (1x3) and (4x1) => Broadcasting ✅
a = np.arange(3).reshape(1,3)
b = np.arange(4).reshape(4,1)

print(a + b)
```

```python
# (2x2) and (1,) => Broadcasting ✅
a = np.arange(4).reshape(2,2)
b = np.arange(1)

print(a + b)
```

```python
# (3x4) and (4,3) => Broadcasting ❌
a = np.arange(12).reshape(3,4)
b = np.arange(12).reshape(4,3)

# print(a + b)
```

```python
# (4x4) and (2,2) => Broadcasting ❌
a = np.arange(16).reshape(4,4)
b = np.arange(4).reshape(2,2)

# print(a + b)
```

>5. Mathematical Operations


```python
# A. Sigmoid Function   -> 1/(1+e ^−x)
def simboid(array):
  return 1 / (1 + np.exp(-array))

arr = np.arange(10)
print(simboid(arr))
```

```python
# B. Mean Squared Error (MSE)
def mse(actual, pedicted):
  return np.mean((actual - pridected)**2 ) 

actual = np.random.randint(1, 50, 25)
pridected = np.random.randint(1, 50, 25)

print(mse(actual, pridected))
```

```python
# C. Working with Missing Values

a = np.array([1,2,3,4,np.nan,6,7,8,9])
print(a)
# Identify missing value
np.isnan(a)
print(a[~np.isnan(a)])
```

>6. Plotting Graphs

```python
# Graph: y = x 
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
y = x
plt.plot(x,y)
```

```python
# Graph: Parabola -> y = x**2 
x = np.linspace(-10,10, 100)
y = x**2

plt.plot(x,y)
```

```python
# Graph: Trigo -> y = sin(x) 
x = np.linspace(-10,10, 100)
y = np.sin(x)

plt.plot(x,y)
```

```python
# Graph: sigmboid -> y = 1/1+ e^-x 
x = np.linspace(-50,50, 100)
y = 1 / 1 + np.exp(-x)

plt.plot(x,y)
```
