###`Q-2:` Replace multiples of 3 or 5 as 0 in the given array.
```
arr=[1 2 3 4 5 6 7 9]

result-> [1 2 0 4 0 0 7 0]
```

````python
arr = np.array([1, 2, 3, 4, 5, 6, 7, 9])
for e in arr:
  if e % 3 == 0 or e % 5 == 0:
    arr[np.where(arr == e )]  = 0
print(arr)    
````

### `Q-9:` Write a program implement Boradcasting Rule to check if two array can be added or not.
Given tuples of shapes.
```
shape of a- (3,2,2)
shape of b- (2,2)

check_broadcast(a, b) -> return Boolean (True if can broadcasted, False other wise.)
```

````python
#  Dimention same
#  any 1 
shape_a = (3,2,2)
shape_b = (2,2)

# if diamentions are same
if len(shape_a) == len(shape_b):
  if shape_a == shape_b:
    print(f'Broadcasting of {shape_a} and {shape_b} are possible ✅')
  
  elif 1 in shape_b:
    index = (shape_b.index(1))
    diff = shape_a[index] - shape_b[index] 
    if shape_b[index] + 2 == shape_a[index]:
      print(f'Broadcasting of {shape_a} and {shape_b} are possible ✅') 
    else:
      print(f'Broadcasting of {shape_a} and {shape_b} are not possible ❌')

# if diamentions are not same
else:
  if len(shape_a) < len(shape_b):
    min_shape = shape_a 
  else:  
    min_shape = shape_b 
  
  x,y = min_shape
  min_shape = (1,x,y)
  
  diff = shape_a[0] - min_shape[0]   
  if min_shape[0] + diff == shape_a[0]:
    print(f'Broadcasting of {shape_a} and {shape_b} are possible ✅')
  else:  
    print(f'Broadcasting of {shape_a} and {shape_b} are not possible ❌')
````
