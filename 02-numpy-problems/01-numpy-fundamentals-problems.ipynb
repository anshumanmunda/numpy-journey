###`Q-19:` Write a function which will accept 2 arguments.
First: A 1D numpy array arr

Second: An integer n {Please make sure n<=len(arr)}

Output: The output should be the nth largest item out of the array
```
# Example1 : arr=(12,34,40,7,1,0) and n=3, the output should be 12
# Example2 : arr=(12,34,40,7,1,0) and n=1, the output should be 40
```

````python
def third_largest(array:np.ndarray, n:int) -> int:

  if n > np.size(array):
    print("n can not be greater than length of array. ")

    exit
  elif not isinstance(array, np.ndarray):
    print("You must pass an numpy array ")
    exit
  
  else:
    temp_list = []
    temp_array = np.array([])
    temp_array = array
    i = 1
    while i<=n:
      max = (np.max(temp_array))
      temp_list.append(max)
      max_index = np.where( temp_array == max )
      
      new_arr = np.delete(temp_array, max_index)
      temp_array = new_arr
      i += 1
  return  temp_list[n-1]


b1 = np.array([12,34,40,7,1,0])
b2 = np.array([12,34,40,7,1,0])
l = [1,2]

result = third_largest(b1, 3)
print(result)
````
